#### **3. Strings**

Un string es simplemente una secuencia de caracteres (letras, números, símbolos). En Python, los strings se escriben entre comillas (simples o dobles).  
Los strings nos permiten trabajar con texto y son fundamentales en muchos programas, como los que hacen tareas de entrada y salida de información.

A continuación tienes varios ejemplos de Strings, ejecuta las celdas para ver el resultado:

1. Crea un string

```python
saludo = "¡Hola Mundo!"

print(saludo)
```

2. Concatena dos strings

```python
nombre = "Juan"
saludo = "Hola, " + nombre  # para esto sirven las variables, el nombre no tiene por qué ser constante

print(saludo)
```

3. Convierte a mayúsculas

```python
texto = "python"

print(texto.upper())  # Imprime "PYTHON"

# ojo, la variable texto NO ha cambiado
# se ha generado una variable NUEVA a partir de texto
# es como texto pero en mayúsculas
# pero texto NO ha cambiado

# texto.upper() devuelve una variable nueva
# no la asigno a ningún nombre de variable, pero la imprimo
```

4. Convierte a minúsculas

```python
texto = "PYTHON"

print(texto.lower())  # Imprime "python"

# lo mismo pero crea una variable NUEVA que está en minúsculas
```

5. Cuenta la longitud de un string

```python
texto = "Python"

print(len(texto))  # Imprime 6
```

6. Accede a un carácter específico del string

```python
texto = "Python"

print(texto[2])  # Imprime "t"

# un texto es una cadena de caracteres, puedo acceder con corchetes y el índice, como hacíamos con las listas
```

7. Reemplaza un texto dentro del string

```python
texto = "Hola, mundo"
nuevo_texto = texto.replace("mundo", "Python")

print(nuevo_texto)  # Imprime "Hola, Python"

# replace es un método de los strings que devuelve un string NUEVO, NO modifica el original
```

8. Verifica si un string contiene una palabra

```python
texto = "Estoy aprendiendo Python"

print("Python" in texto)  # Imprime True
```

9. Elimina los espacios al principio y al final de un string

```python
texto = "  Hola  "

print(texto.strip())  # Imprime "Hola"

# strip tampoco modifica el original
# de hecho, los strings son inmutables, no vais a poder modificarlos una vez creados
# sí podréis crear strings nuevos a partir de strings que ya existen
```

10. Divide un string en una lista

```python
texto = "Python es fácil"
palabras = texto.split()

print(palabras)  # Imprime ['Python', 'es', 'fácil']


# split es un método de strings que separa por el espacio en blanco, por defecto, y devuelve una lista
```
