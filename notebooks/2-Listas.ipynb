#### **2. Listas**

¿Qué es una lista?
Una lista es un tipo de variable que puede almacenar varios valores a la vez. Es como una caja especial que puede contener más de un objeto dentro.  
Los elementos dentro de una lista pueden ser números, textos o incluso otras listas. Las listas son muy útiles cuando necesitas almacenar y trabajar con varios datos relacionados.

A continuación tienes varios ejemplos de Listas, ejecuta las celdas para ver el resultado:

1. Crea una lista de números

```python
numeros = [1, 2, 3, 4, 5]  # usamos corchetes porque es una lista

print(numeros)
```

2. Accede a un elemento de la lista

```python
colores = ["rojo", "verde", "azul"]

print(colores[1])  # Imprime "verde"

# siempre SE ACCEDE a un elemento usando corchetes
# los índices empiezan en 0 en Python
#            0        1       2
#            -3      -2       -1
# Y puedo contar hacia delante o hacia atrás, es lo mismo
```

3. Modifica un elemento de la lista

```python
frutas = ["manzana", "naranja", "kiwi"]
frutas[1] = "pera"

print(frutas)

# accedo con los corchetes y sobrescribo el elemento
# ¡las listas son modificables!
```

4. Agregar un elemento al final de la lista

```python
colores = ["rojo", "verde"]
colores.append("azul")

print(colores)

# las listas tienen un método que se llama append
# una_lista.append(un_elemento)  lo que hace es meter al final de la lista   una_lista   ese un_elemento
# fíjate en el punto entre una_lista y append
# ese punto aparecerá mucho en Python
# le estamos diciendo que tome una_lista y le haga algo, en este caso, un append a ESA una_lista
```

5. Inserta un elemento en una posición específica

```python
colores = ["rojo", "verde"]
colores.insert(1, "amarillo")

print(colores)

# insert es similar a append, pero inserta en el índice que le decimos el elemento que le decimos
# ¿a qué lista? Sí, a la que está antes de del punto y de insert
```

6. Elimina un elemento de la lista

```python
frutas = ["manzana", "naranja", "kiwi"]
frutas.remove("naranja")

print(frutas)

# remove es un método de las listas para eliminar
# ojo, modifica la lista (igual que append también modifica la lista o insert, cada uno sus modificaciones)
```

7. Accede al último elemento de la lista

```python
numeros = [1, 2, 3, 4, 5]

print(numeros[-1])  # Imprime 5
```

8. Cuenta cuántos elementos tiene una lista

```python
colores = ["rojo", "verde", "azul"]

print(len(colores))

# len es una built-in function, quiere decir que es una función que NO necesita un punto delante
# len te dice cómo es de largo lo que le pones dentro de los paréntesis
```

9. Concatena dos listas

```python
lista1 = [1, 2, 3]
lista2 = [4, 5, 6]
lista_completa = lista1 + lista2

print(lista_completa)

# el signo más entre listas, crea una lista nueva que es la concatenación de las anteriores
```

10. Itera sobre los elementos de la lista

```python
colores = ["rojo", "verde", "azul"]

for color in colores:
    print(color)

# Nuestro primer bucle for
# un bucle da vueltas
# aquí quiero que tome cada valor de la lista colores
# lo asocie a color
# entre en en bucle y lo imprima
# si te fijas, el bucle dará tres vueltas
```
