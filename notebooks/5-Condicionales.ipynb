#### **5. Condicionales (if, elif, else)**


¿Qué es una condicional?
Las condicionales son instrucciones que permiten tomar decisiones en un programa. Por ejemplo, podemos decir: "Si el precio es mayor a 50, entonces es costoso; de lo contrario, es económico".

A continuación tienes varios ejemplos de Condicionales, ejecuta las celdas para ver el resultado:

1. Condicional simple

```python
producto = "ordenador"

if producto == "ordenador":
    print("Este es un producto electrónico")

# ojo a los espacios en blanco iniciales

```

2. Condicional con else

```python
producto = "libro"

if producto == "ordenador":
    print("Este es un producto electrónico")
else:
    print("Este es un producto no electrónico")
    
# si entras en el bloque del if, ya no entras en el else

```

3. Condicional con elif

```python
precio = 100

if precio >= 150:
    print("Producto muy costoso")
elif precio >= 50:
    print("Producto moderadamente costoso")
else:
    print("Producto económico")

# if - elif - else 
# elif es lo miso que else - if


# la primera condición que sea cierta, entra en el bloque y no 
# evalúa más por el else

```

4. Verifica si un número es positivo

```python
numero = 7

if numero > 0:
    print("El número es positivo")
else:
    print("El número no es positivo")


```

5. Verifica si un texto está en una lista

```python
productos = ["ordenador", "smartphone", "tablet"]
producto = "tablet"

if producto in productos:
    print("Producto disponible")
else:
    print("Producto no disponible")


```

6. Condicional anidado

```python
tiempo = "lluvioso"
si_no = "salir"

if tiempo == "soleado":
    if si_no == "salir":
        print("Lleva gafas de sol")
    else:
        print("Quédate en casa")


# Realmente no es mucho más difícil si practicamos y aplicamos
# un divide y vencerás --> ¡estúdialo por partes!
```

7. Verifica si un número es divisible entre 3

```python
numero = 9

if numero % 3 == 0:
    print("El número es divisible entre 3")
else:
    print("El número no es divisible entre 3")

# el % con los números es preguntar si divido entre 3, ¿el resto es cero?, es decir, múltiplo de 3

```

8. Comprobación de rango

```python
puntuacion = 75

if 50 <= puntuacion <= 90:
    print("Buena puntuacion")
else:
    print("Puntuacion fuera de rango")

# la condición booleana siempre tiene que acabar valiendo o True o False
# Y esta es una posible condición válida

```

9. Comprobación de varias condiciones con and

```python
cargando = True
autorizado = True

if cargando and autorizado:
    print("Proceso permitido")
else:
    print("Proceso no permitido")


# operaciones de and
```

10. Comprobación de varias condiciones con or

```python
tiene_descuento = False
tiene_puntos = True

if tiene_descuento or tiene_puntos:
    print("Puede usar su descuento")
else:
    print("No puede usar descuento")


# ¡divide y vencerás!
```

**Problema 1:**

Escribe un programa que determine si un artículo es un producto electrónico o no, basado en su nombre.

```python
producto = input("Introduce el nombre del producto: ")

if producto == "ordenador" or producto == "smartphone":
    print("Es un producto electrónico")
else:
    print("No es un producto electrónico")

# ¡A jugar!
```

**Problema 2:**

Escribe un programa que verifique si un número es positivo, negativo o cero.

```python
numero = int(input("Introduce un número: "))

if numero > 0:
    print("El número es positivo")
elif numero < 0:
    print("El número es negativo")
else:
    print("El número es cero")

# ¡Juguemos un poco más!
```
