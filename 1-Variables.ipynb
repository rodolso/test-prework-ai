#### **1. Variables**

¿Qué es una variable?
En programación, una variable es un lugar donde guardamos datos, como números o textos. Puedes pensar en una variable como una "caja" en la que colocamos algo. Podemos poner cualquier tipo de datos en esta caja, y luego podemos usar ese valor más tarde en el programa. Las variables son muy útiles porque te permiten almacenar información y manipularla según sea necesario.

##### Solución Problemas Propuestos

**Problema 1:**

Crea dos variables a y b y asigna valores de tu elección. Luego, imprime la suma, resta, multiplicación y división de ambos.

```python
a = 8
b = 4

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicación:", a * b)
print("División:", a / b)

# print SIEMPRE tiene que imprimir por pantalla lo que le paso dentro de los paréntesis (es para lo que se diseñó)
```

**Problema 2:**

Crea una variable precio y asigna un valor. Si el precio es mayor o igual a 100, imprime "Producto costoso". Si no, imprime "Producto económico".

```python
precio = 120

if precio >= 100:               # los dos puntos es porque comienza un bloque de código
    print("Producto costoso")   # ¡ojo a estos espacios en blanco iniciales! si se cumple el if, entro en el bloque
else:                           # si no, dos puntos, ¡bloque de código!
    print("Producto económico") # entonces imprimiría esto otro
```
