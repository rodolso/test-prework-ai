#### **4. Booleanos**

¿Qué es un valor booleano?

Un booleano es un tipo de dato que solo puede tener dos valores: True (verdadero) o False (falso). Los valores booleanos se usan para tomar decisiones en el código.  
Por ejemplo, si un valor es verdadero, ejecutamos una parte del código, y si es falso, ejecutamos otra.

A continuación tienes varios ejemplos de Booleanos, ejecuta las celdas para ver el resultado:

1. Comparación de igualdad

```python
x = 10
y = 10

print(x == y)  # Imprime True

# ojo, dos iguales pegados es para comparar
# un igual es para asignar
```

2. Comparación de desigualdad

```python
x = 10
y = 5

print(x != y)  # Imprime True

# exclamación seguido de igual es para preguntar si es diferente
```

3. Operador lógico and

```python
es_autorizado = True
es_cliente = True

print(es_autorizado and es_cliente)  # Imprime True

# operación booleana AND es Y --> si todo es cierto (True), el resultado es cierto (True). Si no, no.
```

4. Operador lógico or

```python
es_autorizado = True
es_cliente = False

print(es_autorizado or es_cliente)  # Imprime True

# operación booleana OR es O --> si algo es cierto (True), el resultado es cierto (True). Si no, no.
```

5. Negación con not

```python
es_visible = False

print(not es_visible)  # Imprime True

# Negar True --> False
# Negar False --> True
```

6. Comparación mayor que

```python
x = 15
y = 10

print(x > y)  # Imprime True
```

7. Comparación menor que

```python
x = 5
y = 10

print(x < y)  # Imprime True
```

8. Comparación mayor o igual que

```python
x = 10
y = 10

print(x >= y)  # Imprime True

# Ojo, no hay ningún espacio entre mayor e igual
```

9. Comparación menor o igual que

```python
x = 5
y = 10

print(x <= y)  # Imprime True

# Ojo, no hay ningún espacio entre menor e igual
```

10. Comprobación de existencia

```python
colores = ["rojo", "verde", "azul"]

print("rojo" in colores)  # Imprime True
```
